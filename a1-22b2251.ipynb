```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import scipy.stats
from scipy.stats import norm
```

# Part 1

```python
scipy.stats.dweibull
x = np.linspace(0, 10, 10000)  
cdf = scipy.stats.dweibull.cdf(x, c=2)  
pdf = scipy.stats.dweibull.pdf(x, c=2)  

plt.plot(x, cdf, label='CDF')
plt.plot(x, pdf, label='PDF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.show()

mean = scipy.stats.dweibull.mean(c=2)
variance = scipy.stats.dweibull.var(c=2)

print('mean:', mean,'\nvariance:' ,variance)
```

# Part 2

```python
import scipy.stats as stats
c=2
sample_sizes = [5, 10, 20, 40, 80,200]
sample_std =[]
sample_means = []
for sample_size in sample_sizes:
    means = []
    for _  in range(5000):
        sample = stats.dweibull.rvs(c=2, size=sample_size)
        mean = np.mean(sample)
        means.append(mean)
    sample_means.append(means)
    sample_std.append(np.std(means))

fig, axs = plt.subplots(3, 2, figsize=(8,8))
axs = axs.flatten()
print("The sample mean distribution for different values. c =", c)
for i, sample_size in enumerate(sample_sizes):
    axs[i].hist(sample_means[i], bins=30, density=True)
    axs[i].set_title(f'Sample Size: {sample_size}')
    axs[i].set_xlabel('Sample Mean')
    axs[i].set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.plot(sample_sizes, sample_std)
plt.xlabel('Sample Size')
plt.ylabel('Standard Deviation')
```

As the square root of the sample size increases the standard deviation  decreases. 

This can be seen from the plot above and is hence verified.

```python
sample_size_sqrt = np.sqrt(len(sample_sizes))
sample_means_mean = np.mean(sample_means, axis=1)
z = []
for i in range(len(sample_sizes)):
    z.append((sample_means[i]-sample_means_mean[i])/(sample_std[i]/sample_size_sqrt))
```

```python
col = [ 'b', 'g', 'r', 'c', 'm', 'y']
#for i in range(len(sample_sizes)):
#    plt.hist(z[i], bins=30, density=True, alpha=0.6, color=col[i], label='sqrt(N) * (Z_bar - mu) / S')
fig, axs = plt.subplots(3, 2, figsize=(8, 8))
axs = axs.flatten()
for i, sample_size in enumerate(sample_sizes):
    
    axs[i].hist(z[i], bins=30, density=True, alpha=0.6, color=col[i])
    axs[i].set_title(f'Sample Size: {sample_size}')
    axs[i].set_xlabel('Value')
    axs[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
