```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import scipy.stats
```

# Part 1

```python
scipy.stats.dweibull
x = np.linspace(0, 10, 10000)  
cdf = scipy.stats.dweibull.cdf(x, c=2)  
pdf = scipy.stats.dweibull.pdf(x, c=2)  

plt.plot(x, cdf, label='CDF')
plt.plot(x, pdf, label='PDF')
plt.xlabel('x')
plt.ylabel('Probability')
plt.legend()
plt.show()

mean = scipy.stats.dweibull.mean(c=2)
variance = scipy.stats.dweibull.var(c=2)

print('mean:', mean,'\nvariance:' ,variance)
```

# Part 2

```python
import scipy.stats as stats
for c in range(1,4):
    sample_sizes = [5, 10, 20, 40, 80,200]

    sample_means = []
    for sample_size in sample_sizes:
        means = []
        for _  in range(5000):
            sample = stats.dweibull.rvs(c=2, size=sample_size)
            mean = np.mean(sample)
            means.append(mean)
        sample_means.append(means)

    fig, axs = plt.subplots(3, 2, figsize=(8,8))
    axs = axs.flatten()
    print("The sample mean distribution for different values. c =", c)
    for i, sample_size in enumerate(sample_sizes):
        axs[i].hist(sample_means[i], bins=30, density=True)
        axs[i].set_title(f'Sample Size: {sample_size}')
        axs[i].set_xlabel('Sample Mean')
        axs[i].set_ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```
